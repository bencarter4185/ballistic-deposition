## Plotting Raw Data

```python
def gen_filename(params):
    ## Generate a filename for each item we want to check
    filename = "".join([
        "../data/L" + str(params.l) 
        + "_k" + str(params.k) +
        "_seeds" + str(params.seeds)
        + "_pbc" + str(int(params.pbc))
        + "_iseed" + str(params.iseed)
        + ".csv"
    ])

    return filename
```

```python
# Generate a named tuple -- similar to a struct in Rust
FileParams = namedtuple('FileParams', ['l', 'k', 'seeds', 'pbc', 'iseed'])

# Generate params for our example
params = FileParams(l = 8192, k = 1, seeds = 1, pbc = True, iseed = 0)

# Generate our filename from the params generated
filename = gen_filename(params)
```

```python
# Loads csv and unpack
data_in = np.loadtxt(filename, delimiter = ",")

# vht
v = data_in[:,0]
h = data_in[:,1]
t = data_in[:,2]
```

```python
plt.plot(t,v)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
